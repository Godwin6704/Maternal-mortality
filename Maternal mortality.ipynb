### Histogram of a specific columns

##### We create a histogram to visualize the distribution of a specific column and a heatmap to show correlations between columns.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['RATE_PER_100000_NL'], kde=True)
plt.title('Distribution of RATE_PER_100000_NL')
plt.xlabel('RATE_PER_100000_NL')
plt.ylabel('Frequency')
plt.show()
```

### Correlation Heatmap

```python
CORR = df[['RATE_PER_100000_N', 'RATE_PER_100000_NL']].corr()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(CORR, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

### Boxplot

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['RATE_PER_100000_NL'])
plt.title('Boxplot of RATE_PER_100000_NL')
plt.xlabel('RATE_PER_100000_NL')
plt.show()
```

### Group by specific columns 

```python
grouped_data =df.groupby('GEO_NAME_SHORT')['RATE_PER_100000_NL'].mean()
print(grouped_data)
```

### Pivot Table


```python
pivot_table = df.pivot_table(columns="IND_UUID", index='GEO_NAME_SHORT', values='RATE_PER_100000_NL', aggfunc='mean')
pivot_table
```
